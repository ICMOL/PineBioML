# IO

```python
from package.preprocessing.IO import read_file
from package.preprocessing import impute, bagging
from package.selection.ensemble import selector
```

```python
#from package.GUI.IO import choose_a_file
#input_path = choose_a_file()
k = 100
input_path = "./input/GBM/GBM-abundance_gene_MD-ML.xlsx"
output_path = "./output/GBM/"
export_title = "GBM "+ str(k)
```

```python
parms = {
    " log without scale ":(True, False),
    " log ":(True, True),
    " without scale ":(False, False),
    " ":(False, True)
}
```

```python
proteomics = read_file(input_path).iloc[1:, 44:]
#target = pd.Series(1, index = pd.read_csv("C:\project\TouBu\Target.csv")["PDA"].values, name = "target100")
```

```python
for pars in parms:
    log_domain, scale = parms[pars]

    #x = proteomics.drop(["patientID-radiomics", "Tumor/Normal ", "Withdrawn"], axis = 1)
    #y = proteomics["Tumor/Normal "] == "Tumor"
    
    x = proteomics.drop(["class"], axis = 1).astype("float")
    y = proteomics["class"] == "tumor"
    if not log_domain:
        x = np.power(2, x)

    knn = impute.knn_imputer(threshold = 0.3333, n_neighbor = 5)
    x, y = knn.fit_transform(x, y)

    selectors = selector(scale = scale ,log_domain = log_domain)
    important_feature, importance = selectors.select(x, y, k)

    selectors.kernels["Volcano_p"].plotting()

    #importance["target"] = target

    important_feature.to_csv(output_path+ export_title+ pars+ " important feature.csv")
    importance.to_csv(output_path+ export_title+ pars+ " importance.csv")
```
